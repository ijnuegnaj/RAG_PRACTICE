```python
import os
from google.colab import userdata

os.environ['OPENAI_API_KEY'] = userdata.get('ssu')
```

```python
#모델의 응답을 문자열로 파싱하는 데 사용
#언어 모델의 출력(응답 내용)을 단순한 텍스트 형태로 변환하여 사용할 수 있게 함
from langchain_core.output_parsers import StrOutputParser

#대화형 프롬프트의 양식을 정의하는 클래스
from langchain_core.prompts import ChatPromptTemplate

#OpenAI의 언어모델(GPT계열)을 사용하는 라이브러리
from langchain_openai import ChatOpenAI

#위의 각 라이브러리를 활용하여, 객체를 생성함.
#1. 양식에 구멍 뚫기
template = ChatPromptTemplate.from_template('{topic}에 대한 짧은 농담 만들어봐! 한국 아재개그를 참고하도록 해 ㅋㅋ ')
chat = ChatOpenAI(model='gpt-3.5-turbo', temperature=1)
output_parser = StrOutputParser()
```

```python
#Chain을 LCEL 문법을 사용하여 정의함  | (파이프)
#앞의 결과물 -> 뒤의 함수로 전달
chain = template | chat | output_parser
chain.invoke({'topic':'엘리베이터'})
```

```python
chain2 = template | ChatOpenAI() | StrOutputParser()
chain2.invoke({'topic':'AI'})
```

```python
from langchain_core.runnables import RunnablePassthrough

chain3 = {'topic':RunnablePassthrough()} | template | ChatOpenAI() | StrOutputParser()
chain3.invoke('여름')
```
